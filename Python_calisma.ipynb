```python
import pandas as pd
import numpy as np
```

```python
# Excel'den veri import edildi.

df = pd.read_csv('weather_data.csv')
df.head()
```

```python
#Verinin ilk 5 satırını ve son 5 satırı ekrana yazdırıldı.
df.head(6)

```

```python
df.tail(6)
```

```python
df.describe()
```

```python
#Date, City ve Temperature sütunları seçilerek yeni bir görünüm oluşturun ve ekrana yazdırın.
df_2 = df[['Date', 'City' ,'Temperature']]

df_2
```

```python
#City ve Temperature kolonları birlikte listelendi.
df[[ 'City' ,'Temperature']]
```

```python
#Hava sıcaklığı 30 derecenin üzerinde olan gözlemler listelendi
df[df['Temperature'] > 30]
```

```python
# Şehir olarak sadece "Bursa" olan gözlemler listelendi.

df[df['City'] == 'Bursa']
```

```python
# Şehir "İstanbul" olan VE Nem oranı 60'tan büyük olan kayıtlar listelendi
df[(df['City'] == 'İstanbul') & (df['Humidity']> 60)]
```

```python
# Şehir "Ankara" olan VEYA Sıcaklığı 5 dereceden küçük olan kayıtlar listelendi.
df[(df['City'] == 'Ankara') | (df['Temperature'] < 5)]
```

```python
# Sıcaklık 10°C altında olan veya nemi (Humidity) %70 üzerinde olan veriler filtrelendi.
df[(df['Temperature'] < 10) | (df['Humidity'] > 70)]
```

```python
#Temperature (Sıcaklık) değerine göre büyükten küçüğe (azalan) sıralandı
df_sorted = df.sort_values(by = 'Temperature', ascending = False)
df_sorted.head(11)
```

```python
# Veri neme göre azalan şekilde sıralandı.
df.sort_values('Humidity', ascending = False)
```

```python
#Veri Şehir alanına göre Artan şekilde sıralandı.
df.sort_values('City')
```

```python
#Temperature_F adında yeni bir sütun eklendi.
#Temperature_F, sıcaklığın Fahrenheit cinsinden değerini taşımaktadır.
df['Temperature_F'] = df['Temperature'] * (9/5) + 32
df.head()
```

```python
#"FeelsLike" isminde yeni bir kolon oluşturuldu
# Formül:	FeelsLike = Temperature - (Humidity / 100)

df['FeelsLike'] = df['Temperature'] -(df['Humidity']/100)
df.head( )
```

```python
# Her şehirde (City) kaç adet veri kaydı olduğu hesaplandı.
df.groupby('City')['City'].count()
```

```python
#Şehirlere göre ortalama sıcaklık değerleri hesaplandı.
temp_mean = df.groupby('City')['Temperature'].mean()
temp_mean
```

```python
#En yüksek sıcaklığın olduğu satır (tüm bilgileri) bulundu ve yazdırıldı.
df.loc[df['Temperature'].idxmax()]
```

```python
#En düşük nem oranının olduğu satır bulundu ve yazdırıldı.
df.loc[df['Humidity'].idxmin()]
```

```python
#Ortalama şehir sıcaklıkları excele aktarıldı.
temp_mean.to_excel('sehir_sicakliklari.xlsx')
```
